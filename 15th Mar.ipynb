Q1- Explain the following with an example:
1) Artificial Intelligence
AI refers to the simulation of human intelligence processes by machines, especially computer systems. It involves the creation of algorithms that enable computers to perform tasks that would typically require human intelligence. Examples of AI include speech recognition, image processing, and natural language understanding.

2) Machine Learning
Machine Learning is a subset of AI that focuses on the development of algorithms that allow computers to learn from and make predictions or decisions based on data. Instead of being explicitly programmed, machines are trained to improve their performance over time. For example, a spam email filter can learn to classify emails as spam or not based on historical data.

3) Deep Learning
Deep Learning is a specialized subfield of Machine Learning that uses artificial neural networks to model and process complex patterns in data. These networks consist of layers of interconnected nodes that process and transform data. Deep Learning has achieved remarkable success in tasks like image and speech recognition. An example is training a neural network to identify objects in images.

Q2- What is supervised learning? List some examples of supervised learning:
Supervised learning is a type of Machine Learning where the algorithm learns from labeled training data, making predictions or decisions based on that data. The "supervision" comes from the process of providing the algorithm with correct target outputs during training. Examples include:

Image classification: Training a model to identify objects in images with labeled examples of each object.
Spam email detection: Teaching a model to differentiate between spam and non-spam emails using labeled data.
Regression: Predicting a continuous value, like predicting house prices based on features like size and location.
Q3- What is unsupervised learning? List some examples of unsupervised learning:
Unsupervised learning is a type of Machine Learning where the algorithm learns patterns from unlabeled data, without explicit target outputs. The goal is often to find inherent structures or relationships within the data. Examples include:

Clustering: Grouping similar data points together, like segmenting customers based on their purchasing behavior.
Dimensionality reduction: Reducing the number of variables while retaining important information, useful for visualization.
Anomaly detection: Identifying rare or unusual instances in data, such as fraud detection or manufacturing defects.
Q4- What is the difference between AI, ML, DL, and DS?

AI (Artificial Intelligence) is the broader concept of creating machines that can simulate human intelligence.
ML (Machine Learning) is a subset of AI that focuses on algorithms and models that enable machines to learn from data.
DL (Deep Learning) is a specialized subset of ML that uses neural networks to model complex patterns in data.
DS (Data Science) involves extracting insights and knowledge from data using various techniques, including AI and ML.
Q5- What are the differences between supervised, unsupervised, and semi-supervised learning?

Supervised Learning uses labeled data with target outputs for training.
Unsupervised Learning works with unlabeled data to find patterns or structures.
Semi-Supervised Learning combines both labeled and unlabeled data for training, which can enhance performance when labeled data is scarce.
Q6- What is train, test, and validation split? Explain the importance of each term:
In machine learning, the dataset is often split into three parts:

Training Set: Used to train the model by adjusting its parameters based on the provided data.
Validation Set: Used to fine-tune model hyperparameters and prevent overfitting.
Test Set: Used to evaluate the final model's performance on unseen data, providing an estimate of real-world performance.
Q7- How can unsupervised learning be used in anomaly detection?
Unsupervised learning can be used for anomaly detection by learning the "normal" patterns in data and identifying instances that deviate significantly from these patterns. Techniques like clustering or density estimation can help detect unusual data points that might indicate anomalies or outliers.
