- Here's a brief overview of the advantages and disadvantages of different imputation techniques commonly used to handle missing data: 

1. Mean/Median Imputation:

Advantages:
Simple and easy to implement.
Preserves the overall distribution of the data.
Suitable for numerical data with a normal distribution.
Disadvantages:
Ignores relationships between variables.
Can distort correlations and other statistical measures.
Might not work well if data is not normally distributed.

2. Mode Imputation (for Categorical Data):

Advantages:
Simple and works well for categorical data with frequent modes.
Does not distort the distribution of the data.
Disadvantages:
Ignores relationships between variables.
May not work well if modes are rare or evenly distributed.

3. Forward/Backward Filling:

Advantages:

Simple and suitable for time-series data where missing values are likely consecutive.
Disadvantages:
Ignores relationships between variables.
May not work well for non-sequential data or when missing values are not consecutive.

4. K-Nearest Neighbors Imputation:

Advantages:
Considers relationships between variables.
More accurate than simple imputation methods.
Disadvantages:
Computationally intensive, especially for large datasets.
Sensitive to the choice of the number of neighbors (k).

5. Regression Imputation:

Advantages:
Incorporates relationships between variables.
Can be more accurate when relationships are strong.
Disadvantages:
Requires building predictive models, which might not always be suitable.
Assumes that the relationship between variables is linear.

6. Multiple Imputation:

Advantages:
Accounts for uncertainty in the imputation process.
Provides more accurate estimates and confidence intervals.
Disadvantages:
Can be complex to implement.
Requires multiple imputations and subsequent analysis.

7. Imputation Using Machine Learning Models:

Advantages:
Considers complex relationships between variables.
Potentially provides accurate imputations.
Disadvantages:
Requires substantial data preprocessing and model training.
May not always be justified for small datasets.

To perform exploratory data analysis (EDA) on the wine quality dataset and identify features with non-normal distributions, you can follow these steps using Python and libraries like Pandas, Matplotlib, and Seaborn:
In this code snippet:

We load the dataset using pd.read_csv().
We display the first few rows and calculate basic statistics using describe().
We visualize the distribution of each feature using histograms.
We perform the Shapiro-Wilk normality test for each feature. If the p-value is below a significance level (e.g., 0.05), the distribution is considered non-normal.
The Shapiro-Wilk test checks for normality by comparing the observed distribution to the expected normal distribution. If a feature's distribution is found to be non-normal, you can consider applying transformations to improve normality, such as:

Log Transformation: Apply the natural logarithm to the feature's values.
Square Root Transformation: Take the square root of the feature's values.
Box-Cox Transformation: A more general transformation that can handle both positive and negative values, adjusting the power parameter to achieve normality.
Remember that transforming data should be done thoughtfully, considering the domain knowledge and the specific goals of your analysis. Also, it's important to note that not all statistical analyses require perfectly normal data; some are robust to deviations from normality.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Load the dataset
df = pd.read_csv("C:\Programming\\coding\\Pwskills\Excel files\\wine quality\\winequality-red.csv")

# Display the first few rows of the dataset
df.head()
```

```python
df.describe()
```

```python
# Visualize the distribution of each feature using histograms
plt.figure(figsize=(12, 10))
for col in df.columns:
    plt.subplot(4, 3, df.columns.get_loc(col) + 1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.tight_layout()
plt.show()
```

```python
# Test for normality using Shapiro-Wilk test
normality_results = {}

for col in df.columns:
    stat, p = stats.shapiro(df[col])
    normality_results[col] = {'p-value': p, 'is_normal': p >= 0.05}

# Print the normality results for each feature
for feature, result in normality_results.items():
    print(f"{feature}: p-value = {result['p-value']:.4f}, Is normal? {result['is_normal']}")
```

**Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?**

```python
from sklearn.decomposition import PCA

from sklearn.preprocessing import StandardScaler


# Extract the features (excluding quality) for standardization
X = df.drop(columns=['quality'])

# Create a StandardScaler object
scaler = StandardScaler()

# Standardize the features
X_std = scaler.fit_transform(X)

# Convert the standardized features back to a DataFrame
X_std_df = pd.DataFrame(X_std, columns=X.columns)
X_std_df.head()
```

```python

# Perform PCA
pca = PCA()
pca.fit(X_std_df)

# Calculate the explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate the cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Plot the explained variance ratio
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.grid()
plt.show()

# Find the minimum number of principal components to explain 90% of the variance
min_components = np.argmax(cumulative_explained_variance >= 0.90) + 1
print(f"Minimum number of principal components to explain 90% of the variance: {min_components}")
```
