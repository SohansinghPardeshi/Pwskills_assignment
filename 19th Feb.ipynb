# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a Python web framework that allows developers to build web applications quickly and easily. Flask is a lightweight framework that does not require particular tools or libraries, allowing developers to have more control over their web applications.

- Some advantages of Flask framework include:

Lightweight and Flexible: Flask is a lightweight and flexible framework that allows developers to choose the components they want to use and exclude those they don't need. This gives developers more control over their web applications and allows them to build applications that are tailored to their specific needs.

Easy to Learn: Flask has a simple and intuitive API, which makes it easy for developers to learn and use. Flask is a great framework for beginners who are just starting to learn web development.

Great for Small to Medium-sized Projects: Flask is well-suited for building small to medium-sized web applications. It is not recommended for large, complex projects, but it is perfect for simple web applications or prototypes.

Integrated Testing Support: Flask has built-in support for testing, making it easy for developers to write unit tests and integration tests for their web applications.

Large Community: Flask has a large and active community of developers, which means there are many resources available for learning and troubleshooting issues. Additionally, Flask has a wide range of extensions and plugins that developers can use to extend the functionality of their web applications.

# Q3. What is App routing in Flask? Why do we use app routes?

-->In Flask, an app route is a way to associate a URL endpoint with a function that will be executed when that endpoint is accessed by a client. The process of creating app routes in Flask is known as routing.

Flask's routing system is used to map URL paths to view functions that generate the HTTP response returned to the client. When a client makes a request to a specific URL, Flask uses the routing information to find the appropriate view function and then executes it.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

-->In Flask, the url_for() function is used for URL building. This function takes the name of a view function as its first argument and then generates a URL for that function.

The url_for() function can also take additional arguments that correspond to the variable parts of the URL pattern defined in the view function's route decorator. This allows us to generate URLs with dynamic parameters.
