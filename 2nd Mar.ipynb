Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular plotting library for the Python programming language. It provides a wide range of functionalities for creating high-quality visualizations, making it a powerful tool for data exploration, analysis, and presentation.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line plot: This plot displays data points connected by straight line segments. It is useful for visualizing trends and patterns in data over a continuous range.

2. Scatter plot: Scatter plots are used to display the relationship between two numerical variables. Each data point is represented as a marker on the plot, and the position of the marker is determined by the values of the two variables.

3. Bar plot: Bar plots are commonly used for comparing categorical data. They represent the data as rectangular bars, where the height of each bar corresponds to the value of a particular category.

4. Histogram: Histograms are used to visualize the distribution of a single numerical variable. It divides the variable into bins and displays the frequency or count of data points falling into each bin.

5. Pie chart: Pie charts are circular visualizations that represent data as slices of a pie. Each slice corresponds to a category, and the size of the slice represents the proportion or percentage of data belonging to that category.

These are just a few examples, and Matplotlib offers many more types of plots to cater to various visualization needs.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

A scatter plot is a type of plot used to visualize the relationship between two numerical variables. It is particularly useful for identifying patterns, trends, and correlations in data.

In a scatter plot, each data point is represented as a marker on a coordinate grid, where the x-axis represents one variable and the y-axis represents the other variable. The position of each marker on the plot is determined by the values of the corresponding variables for that data point.

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize= (6,3))
plt.scatter(x,y)
plt.xlabel( 'x - axis ' )
plt.ylabel( 'y - axis ' )
plt.grid()
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

The subplot() function in Matplotlib's Pyplot module is used to create a grid of subplots within a single figure. It allows you to divide the figure into multiple smaller plots, each representing a different visualization or aspect of the data.

```python
plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)

plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)

plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)
plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)

plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot, also known as a bar chart or bar graph, is a type of chart that represents categorical data with rectangular bars. Each bar corresponds to a category, and the length or height of the bar represents the value or frequency associated with that category. Bar plots are commonly used to compare and visualize the distribution, composition, or relationship between different categories.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()
```

```python
plt.barh(company,profit)
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a type of chart that displays the summary statistics of a dataset. It provides a visual representation of the distribution of the data, including the median, quartiles, and any potential outliers. Box plots are commonly used to analyze and compare the spread, skewness, and central tendency of different datasets or variables.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
plt.show()
```
