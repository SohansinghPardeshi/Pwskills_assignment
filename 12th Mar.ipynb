# Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np

blue_count =np.random.randint(10,20)
orange_count =np.random.randint(10,20)
yellow_count =np.random.randint(10,20)
green_count =np.random.randint(10,20)
brown_count =np.random.randint(10,20)
red_count =np.random.randint(10,20)
```

```python
observed_freq = np.array([blue_count, orange_count, green_count, yellow_count, red_count, brown_count])

print(sum(observed_freq))
expected_freq = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20]) * sum(observed_freq)

# H0: data meets the expectations
# H1: data does not meet the expectation


alpha = 0.05
chisquare_cal , p_value = stats.chisquare(observed_freq, expected_freq , alpha)
chisquare_table =  stats.chi2.ppf(1-alpha,  sum(observed_freq) - 1)

if chisquare_cal < chisquare_table :
    print("Accept Null Hypothesis ")
else:
    print("Reject Null Hypothesis")

```

# Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.


```python


sample_proportion = 60 / 500
sample_size = 500
confidence_level = 0.95
critical_value = 1.96  # For a 95% confidence level

# Calculate the standard error
standard_error = np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Calculate the margin of error
margin_of_error = critical_value * standard_error

# Calculate the confidence interval
lower_bound = round(sample_proportion - margin_of_error , 4)
upper_bound = round(sample_proportion + margin_of_error, 4)

print(f"95% Confidence Interval: ({lower_bound}, {upper_bound})")
print(f"95% Confidence Interval: ({lower_bound *sample_size}, {upper_bound*sample_size})")

```

# Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.


```python
import matplotlib.pyplot as plt
df = 10

x = np.linspace(0, 30, 500)

chi2_pdf = stats.chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, chi2_pdf, label=f'Chi-Square Distribution (df={df})')


x_shaded = np.linspace(15, 30, 500)
plt.fill_between(x_shaded, stats.chi2.pdf(x_shaded, df), color='orange', alpha=0.5, label='Chi-Square Statistic = 15')


plt.xlabel('Chi-Square Statistic')
plt.ylabel('Probability Density')
plt.title(f'Chi-Square Distribution with {df} Degrees of Freedom')
plt.legend()


plt.grid()
plt.show()
```

# Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
import scipy.stats as stats

# Given data
sample_mean = 72
sample_std_dev = 10
sample_size = 30
population_mean_h0 = 70

# Step 1: State the null and alternative hypotheses
# Null hypothesis: The population mean is equal to 70 (µ = 70).
# Alternative hypothesis: The population mean is significantly different from 70 (µ ≠ 70).

# Step 2: Calculate the test statistic (t-score)
standard_error = sample_std_dev / (sample_size ** 0.5)
t_score = (sample_mean - population_mean_h0) / standard_error

# Step 3: Determine the degrees of freedom
degrees_of_freedom = sample_size - 1

# Step 4: Find the critical value for the given significance level (0.05) and two-tailed test
alpha = 0.05
critical_value = stats.t.ppf(1 - alpha / 2, df=degrees_of_freedom)

# Step 5: Compare the calculated test statistic with the critical value
if abs(t_score) > critical_value:
    print("Reject the null hypothesis. The population mean is significantly different from 70.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the population mean is significantly different from 70.")

# Print the results
print(f"T-score: {t_score:.4f}")
print(f"Critical value: {critical_value:.4f}")
print(f"Degrees of freedom: {degrees_of_freedom}")
```
