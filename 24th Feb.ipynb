Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
df1 = pd.DataFrame({
    'A': [10, 20, 30], 
    'B': [40, 50, 60], 
    'C': [70, 80, 90]}
    )

def reindexer(dataFrame):
    x = 1
    new_index=[]
    for i in range(len(dataFrame)):
        new_index.append(x)
        x+=2
    dataFrame.index=new_index
    print(dataFrame)

reindexer(df1)
```

Q5. How are DataFrame.size() and DataFrame.shape() different?

The DataFrame.size and DataFrame.shape are both attributes of a pandas DataFrame, but they provide different information about the structure of the DataFrame.

DataFrame.size: It returns the total number of elements in the DataFrame, which is equal to the number of rows multiplied by the number of columns. The DataFrame.size attribute does not take into account the content or values of the DataFrame, but only the total number of elements.

DataFrame.shape: It returns a tuple that represents the dimensions of the DataFrame. The tuple contains two elements: the number of rows and the number of columns. It provides the actual structure of the DataFrame in terms of its rows and columns. The DataFrame.shape attribute gives insight into the size and layout of the DataFrame.

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
df5 = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Sales': [10, 20, 15, 30, 25, 40, 35, 50, 45, 60]
})

def calculate_moving_average(df):
    # Sorting the DataFrame by date if not already sorted
    df = df.sort_values('Date')

    # Calculating the moving average
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df

calculate_moving_average(df5)
```

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.

```python
df6 = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']
})
def add_weekday_column(df):
    # Converting 'Date' column to datetime if not already
    df['Date'] = pd.to_datetime(df['Date'])

    # Adding 'Weekday' column
    df['Weekday'] = df['Date'].dt.day_name()
    
    return df
add_weekday_column(df6)
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
df7 = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-10', '2023-02-15']
})

def select_rows_between_dates(df):
    # Converting 'Date' column to datetime if not already
    df['Date'] = pd.to_datetime(df['Date'])

    # Selecting rows between '2023-01-01' and '2023-01-31'
    selected_rows = df[(df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')]

    return selected_rows

select_rows_between_dates(df7)
```
